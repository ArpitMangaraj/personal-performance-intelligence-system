```python
import pandas as pd
import numpy as np
import os
```

```python
np.random.seed(42)
```

```python
days=1000
```

```python
dates = pd.date_range(start="2022-01-01", periods=days)
```

```python
study_hours = np.clip(np.random.normal(5, 1.8, days), 0, 10)
sleep_hours = np.clip(np.random.normal(7, 1.2, days), 4, 9)
screen_time = np.clip(np.random.normal(4, 1.5, days), 1, 9)
mood = np.clip(np.random.normal(7, 1.5, days), 1, 10).astype(int)
exercise = np.clip(np.random.normal(30, 15, days), 0, 90)
caffeine = np.random.randint(0, 5, days)
expenses = np.clip(np.random.normal(250, 100, days), 50, 700)
```

```python
df = pd.DataFrame({
    "date": dates,
    "study_hours": study_hours,
    "sleep_hours": sleep_hours,
    "screen_time": screen_time,
    "mood": mood,
    "exercise_minutes": exercise,
    "caffeine_cups": caffeine,
    "expenses": expenses
})
```

```python
for i in range(0, days, 250):
    df.loc[i:i+20, "study_hours"] += 2
    df.loc[i:i+20, "sleep_hours"] -= 1
    df.loc[i:i+20, "caffeine_cups"] += 1
```

```python
df["productivity_score"] = (
    df["study_hours"] * 0.5 +
    df["sleep_hours"] * 0.3 +
    df["exercise_minutes"] * 0.05 -
    df["screen_time"] * 0.2
)
```

```python
bad_days = np.random.choice(df.index, 8)

df.loc[bad_days, "sleep_hours"] = 3.5
df.loc[bad_days, "study_hours"] = 1
df.loc[bad_days, "mood"] = 3
```
