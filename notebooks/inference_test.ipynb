```python
import joblib
import pandas as pd

loaded_model = joblib.load("productivity_model.pkl")

sample = pd.DataFrame({
    "study_hours": [6],
    "sleep_hours": [7],
    "screen_time": [3],
    "mood": [8],
    "exercise_minutes": [40],
    "caffeine_cups": [2],
    "expenses": [200]
})

prediction = loaded_model.predict(sample)

print("Predicted Productivity:", prediction[0])
```
